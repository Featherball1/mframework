```python
from mframework.autograd.tensor import Tensor
import numpy as np
```

```python

a = Tensor(np.array([[2.0]]), requires_grad=True)
b = Tensor(np.array([[3.0]]), requires_grad=True)
c = a * b
d = c.exp()

d.backward()
print("a.grad:", a.grad)
print("b.grad:", b.grad)
```
